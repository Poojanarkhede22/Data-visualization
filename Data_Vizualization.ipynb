```python
data = sns.load_dataset('tips')
data
```

```python
# strip plot 
strip_plot = sns.stripplot(y = 'tip',data = data,jitter=True)
plt.ylabel('tips')
plt.show()
```

```python
# grouping wise strip plot
sns.stripplot(x = 'day',y = 'tip',data = data)
plt.xlabel('day')
plt.ylabel('tips')
plt.show()
```

```python
sns.stripplot(x = 'day',y = 'tip',data = data,jitter = False)
plt.xlabel('day')
plt.ylabel('tips')
plt.show()
```

```python
# swarm plot
sns.swarmplot(x = 'day',y = 'tip',data = data)
plt.xlabel('day')
plt.ylabel('tips')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',hue='sex',data=data)
plt.ylabel('tips')
plt.show()
```

```python
# changing orientation
sns.swarmplot(x='tip',y='day',hue='sex',data=data,orient='h' )
plt.ylabel('day')
plt.show()
```

```python
# box and violine plot
```

```python
sns.boxplot(x='day',y='tip',data = data)
```

```python
sns.violinplot(x='day',y='tip',data = data)
plt.ylabel('tips')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x = 'day',y = 'tip',data = data)
plt.ylabel('tips')
plt.subplot(1,2,2)
sns.violinplot(x = 'day',y = 'tip',data = data)
plt.ylabel('tips',size=16,x= 1)
```

```python
# combination plot

sns.violinplot(x='day',y='tip',data = data,inner = None, color = 'lightyellow')
sns.stripplot(x='day', y='tip', data=data,jitter=True,size = 4)
plt.ylabel('tips')
plt.show()
```
